# Testing building snippet datasets

```python
import numpy as np
import seaborn as sns

from bioslds.dataset import RandomSnippetDataset
from bioslds.plotting import FigureManager, show_latent
```

```python
snippet1 = np.exp(-0.5 * np.linspace(-3, 3, 50) ** 2)
snippet2 = -np.exp(-0.5 * np.linspace(-1.7, 1.7, 50) ** 4)
snippet3 = 0.5 * np.sin(np.linspace(-1, 50, 50))
snippets = [snippet1, snippet2, snippet3]
```

```python
with FigureManager(1, len(snippets)) as (_, axs):
    for i, ax in enumerate(axs):
        ax.axhline(0, lw=1, c="k", ls=":")
        ax.plot(snippets[i])
        ax.set_ylim(-1.05, 1.05)
        
        ax.set_title(f"snippet {i}")
```

```python
dataset = RandomSnippetDataset(3, 500, snippets, rng=1, dwell_times=35)
```

```python
with FigureManager(len(dataset), 1) as (_, axs):
    for i, ax in enumerate(axs):
        ax.plot(dataset[i].y)
        show_latent(dataset[i].usage_seq, ax=ax)
```
